## Table model - ccat

```python
Var = 'Primary Energy|Coal'
```

```python
XR_filt = XRraw.sel(Variable=Var)
XR_filt = XR_filt.dropna('ModelScenario', how='any')
ModelScenarios = np.array(XR_filt.ModelScenario)
Models = np.array([XR_filt.ModelScenario.data[i].split('|')[0] for i in range(len(XR_filt.ModelScenario))])
Ccat = np.array(XRmeta.sel(ModelScenario=XR_filt.ModelScenario).Category.data)

modscen_new = []
for i in range(len(Models)):
    m = Models[i]
    c = Ccat[i]
    #if len(np.where(Models == m)[0]) >= 10 and c != 'C8':
    modscen_new.append(ModelScenarios[i])
Models = np.array([modscen_new[i].split('|')[0] for i in range(len(modscen_new))])
Ccat = np.array(XRmeta.sel(ModelScenario=modscen_new).Category.data)
unimodels = np.unique(Models)
unicats = np.unique(Ccat)
XR_filt = XR_filt.sel(ModelScenario = modscen_new)
```

```python
numbers = np.zeros(shape=(len(unimodels), len(unicats)))
for m_i, m in enumerate(unimodels):
    for c_i, c in enumerate(unicats):
            numbers[m_i, c_i] = len(np.where((Ccat == c) & (Models == m))[0])
```

```python
df_nums = pd.DataFrame(numbers, index=unimodels, columns=unicats)
df_nums.to_csv("X:/user/dekkerm/Projects/AR6_Variance/variancedecomposition/Data/ModelvsCat_PrimEnCoal.csv")
df_nums
```
